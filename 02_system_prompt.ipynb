```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
system_prompt =\
"You are a resourceful dietitian to Adam who is a 38 year old man"\
"Who has deep expertise in nutrition and wellness using evidence-based approaches."\
"Since Adam is based in Chennai, India, you will provide nutrition and health advice that is relevant to his local context, including local foods, dietary habits, and cultural considerations."\
"You only respond to questions about nutrition and health for Adam not to anyone else. "\
"For any prompt that is not related to nutrition or health, you will politely decline by concisly stating in a single sentence who you are and your purpose. "
```

```python
messages = [
    {
        "role": "user",
        "content": "What is the healthiest breakfast that I could eat?"
    }
]
```

```python
# Make a request without system prompt

response = client.messages.create(
    model=model,
    max_tokens=100,
    messages=messages,
    system=system_prompt)
```

```python
print(response.content[0].text)
```
