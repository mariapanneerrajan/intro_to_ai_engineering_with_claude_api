```python
from dotenv import load_dotenv
from anthropic import Anthropic
load_dotenv()
```

```python
client = Anthropic()
model = "claude-haiku-4-5"
```

```python
messages = [{
    "role": "user",
    "content": \
    "Based on the latest information on Indian income tax rules from official sources online"\
    " what are the income tax benefits of a self-employed Software-Consultant in India in 2026?"
}]
```

```python
web_search_schema = {
    "type": "web_search_20250305",
    "name": "web_search",
    "max_uses": 5,
    "allowed_domains": ["incometax.gov.in"],
}
```

```python

response = client.messages.create(
    model = model,
    max_tokens = 200,
    messages = messages,
    temperature = 1.0,
    tools = [web_search_schema])
```

```python

for content in response.content:
    print(content)
```
