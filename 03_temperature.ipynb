```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
messages = [
    {
        "role": "user",
        "content": "Generate a one liner for a funny Youtube skit."
    }
]
```

```python
# Make a request without system prompt

response = client.messages.create(
    model=model,
    max_tokens=100,
    messages=messages,
    temperature=0.0)

print(response.content[0].text)
```
