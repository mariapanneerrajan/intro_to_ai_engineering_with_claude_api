```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
messages = [{
    "role": "user",
    "content": "Create a funny poem of how Claude is changing software engineering."
}]
```

```python

response = \
    client.messages.create(
    model = model,
    max_tokens = 1000,
    messages = messages,
    temperature = 1.0)

print(response.content[0].text)
```

```python

stream = client.messages.create(
    model = model,
    max_tokens = 1000,
    messages = messages,
    temperature = 1.0,
    stream = True)

for event in stream:
    print(event)
```

```python
with client.messages.stream(
    model = model,
    max_tokens = 1000,
    messages = messages,
    temperature = 1.0) as stream:
    for text in stream.text_stream:
        print(text, end="")
        pass
    print(stream.get_final_message().content[0].text)
```
